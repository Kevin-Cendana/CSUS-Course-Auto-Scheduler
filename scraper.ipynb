```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
import time

# Set the path to the ChromeDriver executable
service = Service('./chromedriver-mac-arm64/chromedriver')

# Create a new instance of the Chrome driver
driver = webdriver.Chrome(service=service)

# Navigate to the target URL
url = "https://www.csus.edu/class-schedule/fall-2024/CPE"
driver.get(url)

# Wait for the page to fully load (adjust the delay as needed)
time.sleep(5)

# Get the page source (rendered HTML)
html_content = driver.page_source

# Print the HTML content
print(html_content)

# Close the browser
driver.quit()

```

```python
from bs4 import BeautifulSoup

# Set up Selenium and retrieve the HTML content
# ...

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
# Iterate over each table div
table_divs = soup.find_all('div', class_='table')
for table_div in table_divs:
    # Find the <h2> element within the table div
    h2_element = table_div.find('h2')

    # If an <h2> element is found, print its text content
    if h2_element:
        h2_text = h2_element.text.strip()
        print(h2_text)

    # Find all <div> elements within the table div (assuming these represent rows)
    rows = table_div.find_all('div', recursive=False)

    # Iterate over each row
    for row in rows:
        # Find all child elements within the row (assuming these represent cells)
        cells = row.find_all(recursive=False)

        # Iterate over each cell and print its text content
        row_data = [cell.text.strip() for cell in cells]
        print(row_data)
```

```python


# Create a dictionary to store the data
data = {}

# Iterate over each table div
for table_div in table_divs:
    # Find the <h2> element within the table div
    h2_element = table_div.find('h2')

    # If an <h2> element is found
    if h2_element:
        h2_text = h2_element.text.strip()
        data[h2_text] = []  # Initialize an empty list for the h2 text

    # Find all <div> elements with role='row' within the table div
    row_divs = table_div.find_all('div', role='row')
    row_divs = row_divs[1:]

    # Iterate over each row div
    for row_div in row_divs:
        # Find all child elements within the row div (assuming these represent cells)
        cells = row_div.find_all(recursive=False)

        # Extract the days, start time, and end time from the cells
        if len(cells) >= 3:
            days = cells[3].text.strip()
            start_time = cells[5].text.strip()
            end_time = cells[6].text.strip()

            # Create an object with the extracted data
            row_data = {
                'days': days,
                'startTime': start_time,
                'endTime': end_time
            }

            # Append the object to the list for the current h2 text
            if h2_text in data:
                data[h2_text].append(row_data)

# Print the dictionary
for h2_text, row_data_list in data.items():
    print(f"\n{h2_text}:")
    for row_data in row_data_list:
        print(f"  Days: {row_data['days']}, Start Time: {row_data['startTime']}, End Time: {row_data['endTime']}")
```
